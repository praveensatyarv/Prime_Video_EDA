```python
# Import dataset
import pandas as pd
titles_data = pd.read_csv("Datasets/amazon_prime_titles.csv")
titles_data.head()
```

```python
# Check for duplicates
titles_data[titles_data.duplicated()]
```

```python
# Check for Null values
titles_data.isnull()

# Show as heat map
import seaborn as sns
sns.heatmap(titles_data.isnull())
```

Q1. Which year highest number of TV shows and movies were released?

```python
# Count the occurences of individual years 
titles_data['release_year'].value_counts()
```

```python
# Bar graph for the count of TV shows and movies released in different years
titles_data['release_year'].value_counts().plot(kind='bar')
```

Q2. Show the number of TV shows and movies in the dataset.

```python
# Bar graph for the number of TV shows and movies in the dataset
titles_data.groupby('type').type.count()
sns.countplot(titles_data, x = 'type')
```

Q3. Show all the movies released in 2021.

```python
# Filter to display movies released in 2021
titles_data[(titles_data['type'] == 'Movie') & (titles_data['release_year'] == 2021)]
```

Q4. Show TV Shows released in India only.

```python
titles_data[(titles_data['type'] == 'TV Show') & (titles_data['country'] == 'India')].title
```

Q5. Show top 10 directors who gave the highest number of TV shows and movies to Prime Video

Q7. How many movies have PG rating in the United States?

```python
titles_data[(titles_data['type'] == 'Movie') & (titles_data['rating'] == 'PG') & (titles_data['country'] == 'United States')].shape
```

Q8. How many TV shows have 18+ rating and were released after 2018?

```python
titles_data[(titles_data['type'] == 'TV Show') & (titles_data['rating'] == '18+') & (titles_data['release_year'] > 2018)]
```

Q9. What is the maximum duration of a movie on Prime Video?

```python
# Splitting the duration field
titles_data[['minutes', 'unit']] = titles_data['duration'].str.split(' ', expand = True)
# Converting minutes to integer before getting max value
titles_data[titles_data['minutes'].astype(str).astype(int) == titles_data['minutes'].astype(str).astype(int).max()]
```
